```python
import numpy as np
import random 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#creating Payoff matrix
payoff = pd.DataFrame({'Test':[[2.5,2.5],[3.5,5]],
             'Presentation':[[5,3.5],[3.5,3.5]]},
            index = ['Test','Presentation'])
```

```python
def game():
    #opening title 
    print('Game Theory')
    print('Payoff Matrix:') #showing the payoff
    print(payoff)
    
    player = int(input('Which player you want to play with? 1,2,3 or 4?')) #choose players to play with
    n = int(input('How many times you want to play?')) #number of rounds
    score = np.zeros(2) #np array to store score
     
    grade = [[0],[0]]
        
    for i in range(n): #loop started
            # player's choice 
            you = int(input('What is your choice? 0 = Test, 1 = Presentation'))
            #random: 50% probability of each choice
            if (player == 1):
                partner = random.choice([0,1])
            #copycat: 1st game presentation, else copy the players last game
            if (player == 2): 
                if (i == 0): partner = 1
                else: partner = last_you #player's choice in last game
            #'Optimist': mostly choose test       
            if (player == 3):
                partner = random.choice([0,0,0,0,0,0,1,1,0])
            #'Pessimist': mostly chose presentation
            if (player == 4):
                partner = random.choice([1,1,1,1,1,1,0,0,1])
            #Show the chosen options of both player
            if (you==0): 
                    print ('\nYou Chose Test')
            else: print('\nYou Chose Presentation')
            if(partner == 0):
                print ('Your Partner Chose Test\n')
            else: print('Your Partner Chose Presentation\n')
            
            #Update score by adding appropriate payoff value
            score += np.array(payoff.iloc[you][partner]) 
            
            #for first game show the score in payoff
            #then each time add the new scores and divide by 2
            #to make the score out of 5
            if (i > 0): score = score/2 
            
            grade[0].append(score[0])
            grade[1].append(score[1])
            
            #show score
            print("Your grade:",score[0])
            print("Your partner's Grade:", score[1])
            
            plt.plot(range(i+2),grade[0], label='Me')
            plt.plot(range(i+2), grade[1], label='My Partner')
            plt.legend()
            plt.show()

            last_you = you #store the player's choice for copycat
     
    plt.title("Final Grades:")
    plt.plot(range(i+2), grade[0], label='Me')
    plt.plot(range(i+2), grade[1], label='My Partner')
    plt.legend()
    plt.show()
     
```
